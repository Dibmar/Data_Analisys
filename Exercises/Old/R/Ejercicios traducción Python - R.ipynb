## Combining datasets

### Example: US States Data

Merge and join operations come up most often when combining data from different sources. Here we will consider an example of some data about US states and their populations. The data files can be found at http://github.com/jakevdp/data-USstates/:

````python
pop = pd.read_csv('F:\Programacion\1.BOOTCAMP\data\general_dfs\song_views.csv')
areas = pd.read_csv('F:\Programacion\1.BOOTCAMP\data\general_dfs\song_views.csv')
abbrevs = pd.read_csv('F:\Programacion\1.BOOTCAMP\data\general_dfs\song_views.csv')

display(pop.head(), areas.head(), abbrevs.head())
````

````python
pop.head()
````

Junta en un mismo DataFrame la informarción de pop y abbrevs, eliminando la información duplicada.

````python
merged = pd.merge(pop, abbrevs, how='outer',
                  left_on='state/region', right_on='abbreviation')
merged = merged.drop('abbreviation', 1) # drop duplicate info
merged.head()
````

````python
merged.describe()
````

````python
merged.isnull().any()
````

Some of the ``population`` info is null; let's figure out which these are!

````python
merged[merged['population'].isnull()]
````

````python
merged.dropna(inplace=True)
merged.head()
````

````python
final = pd.merge(merged, areas, on='state', how='left')
final.head()
````

````python
final.dropna(inplace=True)
final.head()
````

## Fixed-width binning

### Developer age distribution - qiutar

### Binning based on rounding

Crea una nueva columna en el dataframe, que muestre categorías (bins) como se muestra a continuación:


``` 
Age Range: Bin
---------------
 0 -  9  : 0
10 - 19  : 1
20 - 29  : 2
30 - 39  : 3
40 - 49  : 4
50 - 59  : 5
60 - 69  : 6
  ... and so on
```

````python
coder_survey['Age_bin_round'] = np.array(np.floor(
                              np.array(coder_survey['Age']) / 10.))

coder_survey[['ID.x', 'Age', 'Age_bin_round']].iloc[1071:1076]
````

Definir la variable ID.x como variable de identificación. 

````python
pd.melt(coder_survey, id_vars =['ID.x'], value_vars =['Age', 'Age_bin_round'])
````

### Binning based on custom ranges

Crea una nueva columna en el dataframe que incluya categorías agrupadas (bins) tal y como se muestra a continuación:


``` 
Age Range : Bin
---------------
 0 -  15  : 1
16 -  30  : 2
31 -  45  : 3
46 -  60  : 4
61 -  75  : 5
76 - 100  : 6
```

````python
bin_ranges = [0, 15, 30, 45, 60, 75, 100]
bin_names = [1, 2, 3, 4, 5, 6]

coder_survey['Age_bin_custom_range'] = pd.cut(
                                           np.array(
                                              coder_survey['Age']), 
                                              bins=bin_ranges)

# view the binned features 
coder_survey[['ID.x', 'Age', 'Age_bin_round', 
               'Age_bin_custom_range']].iloc[1071:1076]
````
