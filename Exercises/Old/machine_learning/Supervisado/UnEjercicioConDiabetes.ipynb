# Se trata de hacer regresión lineal

```python
# LIBRERÍAS DE ANÁLISIS
from sklearn.datasets import load_diabetes
import pandas as pd

from sklearn import linear_model
from sklearn.linear_model import LinearRegression

# LIBERÍAS DE VISUALIZACIÓN
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print ('listo')
```

```python
# Datos. Este viene con los datos ya sacaditos
diabetes = (sklearn.datasets.load_diabetes())
print(diabetes.DESCR)
```

```python
# División del dataset y creación del modelo
X = diabetes['data']
y = diabetes['target']


diabetes_model = LinearRegression().fit(X, y)
diabetes_model
```

```python
# División de los dataset en train y test. Dado que esto es aprendizaje supervisado, es necesario dividir los datos en dos

# Con estos aprende
diabetes_data_train = diabetes['data'][-20:]
diabetes_target_train = diabetes['target'][-20:]

# Con estos comprobamos que ha aprendido
diabetes_data_test = diabetes['data']
diabetes_target_test = diabetes['target']
```

```python
# Creación del autómata
# Usando la regresión lineal que hemos creado antes, se le inserta el df_data_train y el df_target_train y se usa un 'fit' (sólo une, un 'fit_transform' une y modifica)

killbot = diabetes_model.fit(diabetes_data_train, diabetes_target_train)
print(f'Intercept: {killbot.intercept_}')
print(f'\nCoefficient:\n{killbot.coef_}')
```

```python
# Comprobando que funciona
# Está viendo cómo ha sido la respuesta utilizando los 10 primeros números como referencia

y_diabetes_data_test = killbot.predict(diabetes_data_test)
y_diabetes_data_test[:10]
```

```python
# Con esto se realiza una comprobación exhaustiva
for i in range(len(y_diabetes_data_test)):
	print("X= %s, Predicted= %s" % (diabetes_data_test[i], y_diabetes_data_test[i]))
```
